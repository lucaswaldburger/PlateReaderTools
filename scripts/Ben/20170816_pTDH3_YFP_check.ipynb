```python
#The YPD list was generated with the routine above.  The SDC lists were then generated from the YPD list
#manually typing in coordinates.  This will be customized for each experiment layout.  
well_list = []
for jj in range(1,6):
    rows_138 = 'BEBE'
    columns_138 = [jj,jj,jj+6,jj+6]
    well_list.append([rows_138[kk]+str(columns_138[kk]) for kk in range(0,4)])

for jj in range(1,6):
    rows_139 = 'CFCF'
    columns_139 = [jj,jj,jj+6,jj+6]
    well_list.append([rows_139[kk]+str(columns_139[kk]) for kk in range(0,4)])

#add in coordinates for AHN321 strain 
well_list.append(['B6','E6','B12','E12'])

#add in coordinates for BMH126 strain 
well_list.append(['D5','G5','D11','G11'])

#add in coordinates for BMH127 strain 
well_list.append(['D6','G6','D12','G12'])

    
well_list = list(chain.from_iterable(well_list))
```

```python
blank = np.mean(OD_data['D4'])

growth_data = []
growth_data.append([OD_data[well]-blank for well in well_list])
growth_data = list(chain.from_iterable(growth_data))                        
                          
growth_data_df = pd.DataFrame(growth_data, index=data_index)

#make columns time points
dt = 15.0/60.0
growth_data_df.columns = growth_data_df.columns*dt


growth_data_df
```

```python
#Average across tech reps and plot all strains in log space

fig, ax = plt.subplots()


#average across technical replicates.
growth_data_df_avg = growth_data_df.mean(level='Strain')
growth_data_df_avg_log = np.log(growth_data_df_avg)/np.log(2)

#Plot growth curves
#only show labels on right side of the plot. 
#growth_data_df_avg.loc[["126","127","321"]].transpose().plot(ax = ax, legend = True)
growth_data_df_avg_log.transpose().plot(ax = ax, legend = True, colormap = "gist_rainbow")
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
